## Operators

#### There are so many operators
- 1.`Arithmetic` : +, - , *, /, //(integer division), %(modulo)
- 2.`Assignment` : =, +=, -=, *=, /=, //=, %=
- 3.`Comparison` : ==, !=, >, <, >=, <=
- 4.`Logical` : and, or, not
- 5.`Bitwise` : &, |, ^, ~, <<, >>

```python
x = 10
y = 4

print("----Arithmetic----- ")
print("x / y = ", x/y)
print("x // y = ", x//y)
print("x * y = ", x*y)
print("x % y = ", x%y)

print("---- Bitwise ----")
print("left shift by 2 bit", x<<2)
print("right shift by 1 bit", x>>1)
```

## Conditions

```python
user_input = input("write an number between 1 and 40 : ")
if int(user_input) <= 1:
    print(f"Your number is {user_input} which is less than or equal to 1")
elif int(user_input)> 1 and int(user_input) < 40:
    print(f"Your number is {user_input} which is between 1 and 40")
else:
    print(f"Your number is {user_input} which is greater than or equal to 40")
```

## Strings

```python
name = "Brij"

# slicing
print(name[:2])

# reversing the string
print(name[::-1])

# adding two strings
print(name + ' ' + 'Patel')

# upper case
print(name.upper())

# lower case
print(name.lower())

# removing extra spaces
print("Brij    Patel     ".strip())

# spliting the text
print("brij patel".split(" "))
```
