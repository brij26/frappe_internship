```python
# let's write three functions 
# 1. for adding items to the cart or invoice
# 2. for removing items from the cart or invoice
# 3. to calculate the total cost of invoice
from typing import Dict
```

```python
# 1. add
def add_items_to_invoice(item_name:str, price: float, invoice : Dict[str, float]) -> Dict[str, float]:
    """
    This function takes item name and price of that item and add it to the invoice
    input: {item_name, price, invoice}
    output: {invoice}
    """
    invoice[item_name] = price
    return invoice
```

```python
def remove_item_from_invoice(item_name : str, invoice : Dict[str, float]) -> Dict[str, float]:
    """
    This function takes item name and invoice and remove the item from that invoice
    input: {item_name, invoice}
    output: {invoice}
    """
    if invoice is None:
        return {}
    invoice.pop(item_name, None)
    return invoice
```

```python
def cost_of_invoice(invoice : Dict[str, float]) -> float:
    """
    This function takes invoice and returns total cost of the invoice
    input: {invoice}
    output: {cost}
    """
    cost = 0.0
    for _, price in invoice.items():
        cost += price
    return cost
```

```python
items_available = {"pizza" : 100.00,
         "milk" : 36.00,
         "bread" : 40.00,
         "icecreame" : 55.00}
```

```python
invoice = {}

while True:
    print("\n===== Invoice Menu =====")
    print("1. Add Item")
    print("2. Remove Item")
    print("3. View Invoice")
    print("4. Exit")

    choice = input("Enter your choice: ")

    if choice == "1":
        print("\nAvailable Items:")
        for item, price in items_available.items():
            print(f"{item} : ₹{price}")

        item_name = input("Enter item name: ")

        if item_name in items_available:
            invoice[item_name] = items_available[item_name]
            print(f"{item_name} added successfully.")
        else:
            print("Item not available.")

    elif choice == "2":
        if not invoice:
            print("\nInvoice is empty. Nothing to remove.")
        else:
            print("\nCurrent Invoice:")
            for item, price in invoice.items():
                print(f"{item}: ₹{price}")

            item_name = input("\nEnter the item to remove: ")

            if item_name in invoice:
                invoice = remove_item_from_invoice(item_name, invoice)
                print(f"{item_name} removed successfully.")
            else:
                print("Item not found in the invoice.")

    elif choice == "3":
        print("\nCurrent Invoice")

        if len(invoice) == 0:
            print("Invoice is empty.")
        else:
            total = 0

            for item, price in invoice.items():
                print(f"{item} : ₹{price}")
                total += price

            print(f"Total = ₹{total}")

    elif choice == "4":
        print("Thank you!")
        break

    else:
        print("Invalid choice.")
```
